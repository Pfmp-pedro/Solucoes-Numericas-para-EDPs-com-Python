# Solução por Sistema Linear
O objetivo é montar um sistema com as equações 
de diferenças vindas da aproximação das derivadas por diferenças progressivas e regressivas. Para um único ponto $T_{i, j}$ na malha temos a equação:

$$ \begin{equation}
    - 4T_{i,j} + T_{i+1,j} + T_{i-1,j} + T_{i,j+1} + T_{i,j-1}  = 0
\end{equation}$$

 Que relaciona sua temperatura com as dos pontos ao seu redor.  Numa malha com índices começando em 0, na primeira linha de incógnitas ($i = 1$) tem-se:
 
 
 \begin{align}
        - 4T_{1,1} + T_{2,1} + T_{0,1} + T_{1,2} + T_{1,0}  &= 0 \rightarrow \text{Linha 1, Coluna 1}\\
       - 4T_{1,2} + T_{2,2} + T_{0,2} + T_{1,2} + T_{1,0}  &= 0 \rightarrow \text{Linha 1, Coluna 2}\\
       ... ~~&=~~ ... \\
        - 4T_{1,n-2} + T_{2,n-2} + T_{0,n-2} + T_{1,n-1} + T_{1, n-3}  &= 0\rightarrow \text{Linha 1, Coluna } n-2\\
        - 4T_{2,1} + T_{3,1} + T_{1,1} + T_{2,2} + T_{2,0}  &= 0 \rightarrow \text{Linha 2, Coluna 1}\\
        ... ~~&=~~ ... \\
        ... ~~&=~~ ... \\
        - 4T_{n-2,1} + T_{n-1,1} + T_{n-2,1} + T_{n-2,2} + T_{n-2,0}  &= 0 \rightarrow \text{Linha } n-2\text{, Coluna 1}\\     
        ... ~~&=~~ ... \\
        - 4T_{n-2,n-2} + T_{n-1,n-2} + T_{n-3,n-2} + T_{n-2,n-1} + T_{n-2,n-3}  &= 0 \rightarrow \text{Linha } n-2\text{, Coluna n-2}\\   
\end{align}

Como as condições de cortorno são para temperaturas constantes, os pontos com índices $0$ ou $n - 1$ sempre são conhecidos, então podemos isolá-los do lado direito. Esses pontos só aparecerão quando analisamos a segunda (índice 1) ou penultima (índice $n-2$) linha ou coluna.
Reescrevendo:


 \begin{align}
        - 4T_{1,1} + T_{2,1}  + T_{1,2}   &= -(T_{1,0}+ T_{0,1})                    & \rightarrow \text{Linha 1, Coluna 1}\\
       - 4T_{1,2} + T_{2,2}  + T_{1,2}   &= -( T_{0,2})                                   & \rightarrow \text{Linha 1, Coluna 2}\\
       ... ~~&=~~ ... \\
        - 4T_{1,n-2} + T_{2,n-2}  + T_{1, n-3}  &= -(T_{0,n-2} + T_{1,n-1}) & \rightarrow \text{Linha 1, Coluna } n-2\\
        - 4T_{2,1} + T_{3,1} + T_{1,1} + T_{2,2}   &= -(T_{2,0})                    & \rightarrow \text{Linha 2, Coluna 1}\\
        ... ~~&=~~ ... \\
        ... ~~&=~~ ... \\
        - 4T_{n-2,1} + T_{n-1,1} + T_{n-2,1} + T_{n-2,2}  &= -( T_{n-2,0}) &\rightarrow \text{Linha } n-2\text{, Coluna 1}\\     
        ... ~~&=~~ ... \\
        - 4T_{n-2,n-2}  + T_{n-3,n-2}  + T_{n-2,n-3}  &= -(T_{n-2,n-1} + T_{n-1,n-2}) & \rightarrow \text{Linha } n-2\text{, Coluna n-2}\\   
\end{align}


## Simplificação

![scale=.5](https://i.ibb.co/9bm1FgL/Malha-Basica.png)

Se simplificarmos e analizamos o caso mais simples, onde há apenas um termo que não tem nenhuma condição de contorno ao seu redor, temos:


 \begin{align}
        - 4T_{1,1} + T_{2,1}  + T_{1,2}                    &= -(T_{1,0}+ T_{0,1})                   \\
       - 4T_{1,2} + T_{2,2} + T_{1,1}  + T_{1,2}                    &= -( T_{0,2})                                  \\
        - 4T_{1,3} + T_{2,3}  + T_{1, 2}                  &= -(T_{0,3} + T_{1,4}) \\
        - 4T_{2,1} + T_{3,1} + T_{1,1} + T_{2,2}    &= -(T_{2,0})                    \\
        - 4T_{2,2} + T_{3,2} + T_{1,2} + T_{2,3} + T_{2, 1}   &= 0\\
        - 4T_{2,3} + T_{2,2} + T_{3,3} + T_{1,3} &= -(T_{2,4})\\
        - 4T_{3,1} + T_{3,2}  + T_{2,1}  &= -(T_{3,0}+T_{4,1}) \\     
         - 4T_{3,2} + T_{3,1}  + T_{3,3} + T_{2,2} &= -(T_{4,2}) \\
        - 4T_{3,3}  + T_{3,2}  + T_{2,3}  &= -(T_{3,4} + T_{4,3})   
\end{align}


Essas equações podem ser representadas em forma matricial como:

$$
\begin{bmatrix}
-4 & 1 &  & 1\\
1 & -4 & 1 &  & 1\\
 & 1 & -4 & 0 &  & 1\\
1 &  & 0 & -4 & 1 &  & 1\\
 & 1 &  & 1 & -4 & 1 &  & 1\\
 &  & 1 &  & 1 & -4 & 0 &  & 1\\
 &  &  & 1 &  & 0 & -4 & 1\\
 &  &  &  & 1 &  & 1 & -4 & 1\\
 &  &  &  &  & 1 &  & 1 & -4
\end{bmatrix}
\begin{Bmatrix}
T_{1,1}\\
T_{1,2}\\
T_{1,3}\\
T_{2,1}\\
T_{2,2}\\
T_{2,3}\\
T_{3,1}\\
T_{3,2}\\
T_{3,3}
\end{Bmatrix}=
\begin{Bmatrix}
-(T_{1,0}+T_{0,1})\\
-(T_{0,2})\\
-(T_{0,3}+T_{1,4})\\
-(T_{2,0})\\
0\\
-(T_{2,4})\\
-(T_{3,0}+T_{4,1})\\
-(T_{4,2})\\
-(T_{3,4}+T_{4,3})
\end{Bmatrix}
$$

E de forma compacta:

$$\mathbf{A}\{x\} = \{b\}$$



Onde $\mathbf{A}$ é a matriz dos coeficientes, $\{x\}$ é o vetor das incógnitas e  $\{b\}$ o vetor dos termos independentes. O  vetor $\{x\}$ foi montado percorrendo a malha de linha em linha, assim a matriz $\mathbf{A}$ tem esse aspecto para essa configuração específica.

É fácil perceber que em uma malha maior os $n$ primeiros termos do vetor $\{x\}$ terão condições de contorno, assim como os $n$ últimos e entre esses blocos de valores apenas o primeiro e o último ponto de cada linha terá. Qualquer outro ponto não terá valores conhecidos ao seu redor, assim o lado direito das suas equações de diferenças será sempre zero.


## Implementação

O Bloco abaixo cria a matriz 'T' que conterá as temperarturas. Num primeiro estágio são adicionadas as condições de contorno (apenas condições de temperatura constante até o momento). Essa condições são então manipuladas  para um vetor 'b', que será o vetor dos termos independentes.

Nos blocos seguintes são definidas as funções ``coef_T_const()`` e ``solucao_solve()``. A primeira monta a matriz dos coeficientes para uma malha quadrada com ``n`` linhas. A segunda a soluciona, também adicionando as condições de contorno na matriz resultante.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import time


# Número de linhas e colunas da malha 
n = 70
T = np.zeros([n,n])

# Condições de Contorno
T[0, :] = x_inf = 0 #inf
T[-1, :] = x_sup = 100 #sup
T[:, -1] = x_dir = 50 #dir
T[:, 0] = x_esq = 75 #esq




#Y = np.ravel(X)
#print('Condições de contorno: \n',X)

# C é uma matriz auxiliar, que fornecerá a soma dos valores
# ao redor de cada ponto da linha 1 e n-2 e coluna 1 e n-2
# Baseado nela será gerado o vetor dos termos independentes b
C = np.zeros([n, n])
for i in range(n):
    for j in range(n):
        if i == 1:
            C[i][j] += T[i-1][j]
        if i == n-1: 
            C[i-1][j] += T[i][j]
           
        if j == 1:
            C[i][j] += T[i][j-1]
        if j == n-1: 
            C[i][j-1] += T[i][j]


# Recorte de C
K = C[1:n-1,1:n-1]

# Revel de -K: vetor dos termos independentes
b = np.ravel(-K)


#print('\n Soma das condições de contorno: \n', K)
#print('\n Vetor dos Termos Independentes: \n', b)


# Plot
fig, ax = plt.subplots()
pcolor(T, cmap = 'jet')
cbar = colorbar()
plt.title('Temperatura nos Contornos', fontsize=18)
cbar.ax.set_ylabel('Temperatura', fontsize=12)
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$y$', fontsize=15)
plt.show()


print("Solução por sistema de equações:")

# cronômetro
t1 = time.clock()

# coeficientes
A = coef_T(n)
t12 = time.clock()

t2 = time.clock()
# solução por solve
T = solucao_solve(A, b, n, x_esq, x_dir, x_sup, x_inf)

t22 = time.clock()

print("Tempo de montagem (clock): ", np.round(t12-t1,4), "segundos")
print("Tempo de solução (clock): ", np.round(t22-t2,4), "segundos")

fig, ax = plt.subplots()
pcolor(T, cmap = 'jet')
cbar = colorbar()
plt.title('Distribuição de Temperatura', fontsize=18)
cbar.ax.set_ylabel('Temperatura', fontsize=12)
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$y$', fontsize=15)
plt.show()

print (len(T[0]))
```

# Montagem da Matriz dos Coeficientes


```python
import numpy as np

def coef_T (n):
    '''Esta função monta a matriz dos coeficientes de um 
    sistema de equações de diferenças problemas do tipo:
    laplaciano de T = 0, cond. de cont. com temperatura constante, malha quadrada.
    Parâmetro: n = número de linha na malha'''
    n_linhas = n                       # número de linhas/colunas  na matriz das temperaturas
    N = n_linhas-2                     # número de linhas/colunas na matriz das incógnitas

    X = np.zeros([N**2,N**2])          # montagem da matriz das incógnitas
    Mult_N = np.array([N*i for i in np.arange(1, N**2 +1, 1)])
    k = [N*i for i in np.arange(1, N**4, 1)]
    for i in range(N**2): #[OK]
        for j in range(N**2): #[OK]
            if i == j:
                aux = int(N*i)
                #print(Mult_N+1)
                X[i][j] = -4               # Diagonal Principal
                if i < N**2-1:             # Coluna da direita: se a linha i não é a ultima, o vizinho da direita vale 1
                    if j+1 not in Mult_N:  # Se a coluna da direita não é um multiplo de N ela vale 1
                        X[i][j+1] = 1

                if i > 0:                  # Coluna da esquerda: se a linha i não é a primeira
                    if j-1 not in Mult_N-1: # Se a coluna da esquerda não é um a menos de um multiplo de N ela vale 1
                        X[i][j-1] = 1 

                if j+N <= N**2-1:          # N-ésima coluna a partir da Diag. Princ.: se a coluna da direita não é a ultima, ela vale 1
                    X[i][j+N] = 1  
                if j-N >= 0:               # N-ésima coluna a partir da Diag. Princ.:
                    X[i][j-N] = 1
    return X
```

# Solução por solve:

```python
import numpy as np
from scipy.linalg import solve
import matplotlib.pyplot as plt
import time

def solucao_solve(A, b, n, x_e, x_d, x_s, x_i):
    '''Parâmetros: 
    A = matriz dos coeficientes
    b = vetor dos t. independentes
    n = número de linhas da malha
    x_e = temperatura no lado esquerdo
    x_d = // lado direito
    x_s = // lado superior
    x_i = // lado inferior

    Retorno: matriz y
    '''

    N = n-2

    # Solução por solve()
    y = solve(A,b)

    # Tranformando y em uma matriz NxN
    y = y.reshape((N,N))

    # Expandido y para (N+2)x(N+2) com as codições de contorno
    # nas primeiras e ultimas linhas e colunas
    y = np.insert(y, 0, x_e, axis=1)
    y = np.insert(y, n-1, x_d, axis=1)
    y = np.insert(y, n-2, x_s, axis=0)
    y = np.insert(y, 0, x_i, axis=0)

    return y
```

### Decomposição LU para Matrizes Pentadiagonais


```python
# d, e, f = LEdecomp5(d, e, f)
# Decomposição LU de uma matriz pentadiagonal [a], onde 
# {f}, {e} e {d} são as diagonais de [a]. 
# Na saída {f}, {e} e {d} são as diagonais da matriz decomposta

# x = LUsolve5(d, e, f, b)
# Resolve [a]{x} = {b}, onde {f}, {e} e {d} são os vetores retornados 
# da decomposição.


def LUdecomp5(d, e, f):
    n = len(d)
    for k in range(n-2):
        lam = e[k]/d[k]
        d[k+1] = d[k+1] - lam*e[k]
        e[k+1] = e[k+1] - lam*f[k]
        e[k] = lam
    
        lam = f[k]/d[k]
        d[k+2] = d[k+2] - lam*f[k]
        f[k] = lam

    lam = e[n-2]/d[n-2]
    d[n-1] = d[n-1] - lam*e[n-2]
    e[n-2] = lam
    return d, e, f



def LUsolve5(d, e, f, b):
    n = len(d)+1
    b[1] = b[1] - e[0]*b[0]
  
    for k in range(2, n):
        b[k] = b[k] - e[k-1]*b[k-1] - f[k-2]*b[k-2]
    b[n-1] = b[n-1]/d[n-1]
    b[n-2] = b[n-2]/d[n-2] - e[n-2]*b[n-1]
  
    for k in range(n-3, -1, -1):
        b[k] = b[k]/d[k] - e[k]*b[k+1] - f[k]*b[k+2]
  
    return b
    
print('**** RODADO! ****')
```

```python
import numpy as np
import matplotlib.pyplot as plt
# from LUdecomp import *

# m = número de equações, começa em zero

# Função para as equações de diferenças finitas
# Define os valores das diagonais da matriz dos coeficientes 
# e a metriz dos termos independentes a serem usadas na decomposição 

def equations(x, h, m): 
  
    h4 = h**4
  
    # Diagonais
    d = np.ones(m+1)*6.0 # Diagonal principal
    d[0] = 1.0
    d[1] = 7.0
    d[m-1] = 7.0
    d[m] = 3.0
  
    e = np.ones(m)*(-4.0) # Diagonal antes de principal
    e[0] = 0.0
  
    f = np.ones(m-1) # Segunda diagonal antes da princ.
    f[0] = 0.0
  
    # Termos independentes
    b = np.zeros(m+1)
    b[m] = 0.5*h**3
  
    return (d, e, f, b)

xStart = 0.0 # x na direita
xStop = 0.5 # x na esquerda

m = 20 
h = (xStop - xStart)/m # passo da malha
x = np.arange(xStart, xStop + h, h) # eixo x

# Equações
d, e, f, b = equations(x, h, m)
# Solução
d, e, f = LUdecomp5(d, e, f)
y = LUsolve5(d, e, f, b)


# Impressão dos valores
print('\n        x              y')
print('{:14.5e} {:14.5e}'.format(x[m-1], y[m-1]))
print('{:14.5e} {:14.5e}'.format(x[m], y[m]))

#print('\n', x)
#print('\n', y)

plt.plot(x, y)

```
