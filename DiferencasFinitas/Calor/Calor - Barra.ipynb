### Parabólica
Método explícito
$$ T^{l+1}_{i} = T^{l}_{i} + \lambda(T^{l}_{i+1} - 2T^{l}_{i} + T^{l}_{i-1}) $$
$$ \lambda = k \cfrac{\Delta t}{(\Delta x)^2} $$
$ l $ = instante
$ i $ = posição

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import matplotlib


# Passos
tf = 13
dt = .1
dx = .5
k = 0.49/(2.7*0.2174)

lamb = (k * dt) / (dx)**2
print (lamb)

# Tamanho da barra
L = 10

# Vetor da temperatura
X = np.arange(0, L+dx, dx)
Inst = np.arange(0, tf, dt)

print(len(Inst))
print(len(X))

T = np.zeros([len(Inst), len(X)])


# primeira posição: instante, segunda temperatura



#################   CONDIÇÃO INICIAL   #################
#for i in range(1,len(X) -1):
#    T[0,i] = 2*X[i]
    
#################   CONDIÇÕES DE CONTORNO   #################
T[:, 0] = 100 # Esquerda
T[:, -1] = 50 # Direita




print("T inicial")
print(T[0,:])

################   SOLUÇÃO   ###################

for l in range(1, len(Inst) -1):
        for i in range(1,len(X) -1):
            T[l,i] = T[l-1,i] + lamb*(T[l-1,i+1] -2*T[l-1,i] + T[l-1,i-1])
            #print("i",i)
            
            #print ("--- Posição ---",i)
        #print("l",l)
        #print("--- Instante ---",l)
        #print(T[l,:])



            
#print("T depois")
#print(T)


####################   PLOTS   #######################

fig, ax = plt.subplots() # Cria a figura com um subplot
pcolor(T, cmap='jet')
cbar = colorbar()


cbar.ax.set_ylabel('Temperatura [°C]', fontsize=16)
ax.set_ylabel('Tempo [x 10 s]', fontsize=16)
#savefig('Gradiente_dt=0.5cm.pdf')
show()


plt.plot(X,T[30,:],X,T[60,:],X,T[90,:],X,T[120,:])
plt.legend(('3s','6s','9s','12s'))
plt.xlabel('x [cm]', fontsize=16)
plt.ylabel('Temperatura [°C]', fontsize=16)
#savefig('Linhas_dt=0.5cm.pdf')

######################################################################
```
